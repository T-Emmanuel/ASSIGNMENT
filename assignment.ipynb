```python
from sklearn.preprocessing import MinMaxScaler #importing necesary libraries
scaler = MinMaxScaler() # Initialize the Min-Max Scaler
df['Normalized_Sales'] = scaler.fit_transform(df[['Sales']]) #Normalize the 'Sales' column
print(df[['Sales', 'Normalized_Sales']])
```

```python
#CATEGORIZE SALES LEVELS
#DEFINING THE BIN AND LABEL
df= pd.read_csv('sales_data.csv')
bins = [0, 200, 500, float('inf')]  # (Low, Medium, High)
labels = ['Low', 'Medium', 'High']
#CREATE A NEW COLUMN CATEGORIZING CELLS
df['Sales_Category'] = pd.cut(df['Sales'], bins=bins, labels=labels, right=False)
#PRINT RESULT
print(df)
```

## Part 6: Data Visualization Principles

### Instructions:
1. Create the following visualizations:
   - A bar chart of total sales per region.
   - A line chart showing monthly sales trends.
   - A scatter plot of Sales vs. Profit, with points colored by category.

2. Interpret the insights from each visualization and describe how they might inform business decisions.


```python
#CREATING a bar chart of total sales per region
import matplotlib.pyplot as plt #importing data visualization library

region_sales = df.groupby('Region')['Sales'].sum() #Grouping data by 'Region' and calculating total sales

# Plotting the bar chart
region_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales per Region', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()

plt.show() #Displaying the plot
```

```python
# Converting 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extracting month and year from 'Order Date'
df['Month_Year'] = df['Order Date'].dt.to_period('M')

# Counting the number of orders per month
monthly_sales = df.groupby('Month_Year')['Order Date'].count()

# Plotting the line chart of monthly sales trend
monthly_sales.plot(kind='line', marker='o', color='green', figsize=(7.5, 4.5)) 
plt.title('Monthly Sales Trends', fontsize=14)
plt.xlabel('Month-Year', fontsize=12)
plt.ylabel('Order Count', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()

plt.show() # Display the plot
```

```python
# A scatter plot of Sales vs. Profit, with points colored by category.
#Creating the scatter plot
plt.figure(figsize=(6, 4))
scatter = plt.scatter(df['Sales'], df['Profit'], c=pd.Categorical(df['Category']).codes, cmap='viridis') # Color points by Category using the viridis color map
plt.colorbar(scatter, label='Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs. Profit colored by Category')
plt.show() 
```

### insights from the visualizations:
 1: Regions with higher total sales indicate stronger performance in those areas. If some regions have significantly lower sales, it might highlight opportunities for improvement, such as targeted marketing strategies or operational adjustments.

 
 2:Peaks and valleys in the trend show sales fluctuations. Seasonal patterns or promotions might explain higher sales in certain months. November has the lowest sales and the reason for the decline should be highly investigated and taken care of.

 
 3:This visualization helps identify if high sales directly correspond to high profits, and whether specific categories outperform others.
